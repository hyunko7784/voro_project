#Get text from web (Web Scraping)

```python
def get_text(link):
  req=r.get(link)
  soup=BeautifulSoup(req.text,'html.parser')
  texts=[]
  seen=set()
  specs=soup.find_all('div',class_='shg-c-lg-3 shg-c-md-3 shg-c-sm-3 shg-c-xs-6')
  if specs and link!=mantisprose_web:
    for div in specs:
      spans=div.find_all('span')
      for span in spans:
        text=span.text
        if text not in seen:
          texts.append(span.text)
          if text not in ['Yes','No','TBD','N/A','Front/Rear']:
            seen.add(text)
  else:
    specs=soup.find_all('div',class_='row')
    if specs:
      for div in specs:
        inputs=div.find_all(class_='inputs')
        for input in inputs:
          text=input.text
          if text not in seen:
            texts.append(text)
            if text not in ['Yes','No','TBD','N/A','Front/Rear','(2021) No | (2022) Yes']:
              seen.add(text)
    else:
      specs=soup.find_all('div',class_='shg-rich-text shg-theme-text-content')
      for div in specs:
        li=div.findAll('li')
        if li:
          for i in li:
            text=i.text
            if text not in seen:
              texts.append(text)
              if text not in ['Yes','No','TBD','N/A','Front/Rear']:
                seen.add(text)
  return texts
```

```python
def dic_web(webs):
  lis={}
  for web in webs:
    title=web.split('/')[-1]
    text=get_text(web)
    if (web==eaglepro_web):
      new_dic={j[0]:j[1] for j in [i.split(':') for i in text[0].split('\n') if i]}
      lis[title]=new_dic
    elif (web==victor_web):
      d={}
      for i in text:
        j=i.split(':')
        try:
          d[j[0]]=j[1]
        except:
          pass
      lis[title]=d
    else:
      if web==mantiskinggt_web:
        text=text[:-17]
      elif web==xlimited_web:
        del text[-2]
        del text[-2]
      elif web==wolfkinggt_web:
        text=text[:-19]
      elif web==roadrunner_web:
        text=text[:-8]
      elif web==wolfwarriorxgt_web:
        text=text[:-20]
      elif web==wolfwarrior_web:
        text=text[:-19]
      d={}
      try:
        for i in range(len(text))[::2]:
          d[text[i]]=text[i+1]
      except:
        d=title
      lis[title]=d
  return lis
```

```python
def combine_dic(lis_dic):
  dd=defaultdict(list)
  for d in tuple(lis_dic):
    for key, value in sorted(d.items()):
      dd[key].append(value)
  return dd
```

```python
def sep_char(dimen,char):
  i=0
  while dimen[i]!=char:
    i+=1
  return dimen[:i],dimen[i:]
def folded(row):
  return sep_char(row,'U')[0]
def unfolded(row):
  return sep_char(row,'U')[1]
```

```python
df['DimensionsFolded']=df['Dimensions'].apply(folded)
df['DimensionsUpright']=df['Dimensions'].apply(unfolded)
df.drop(columns='Dimensions',inplace=True)
```
